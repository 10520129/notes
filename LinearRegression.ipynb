# Linear Regression

*  $ \hat{y} = \beta_0 + \beta_1 * \hat{x} $

* $ y_i =  \beta_0 + \beta_1 * x_i + error $

* $ Loss(\beta_0, \beta_1) = \sum\limits_{i=1}^{n} (y_i - \hat{y} ) ^ 2 $

* 偏微分-> $ \beta_0 = \bar{y} - \beta_1 * \bar{x} $
* 偏微分 + 上式-> $ \beta_1 = \frac {\sum\limits_{i=1}^{n} (y_i - \bar{y}) (x_i - \bar{x})} {\sum\limits_{i=1}^{n} (x_i - \bar{x})^2} $

```python
import matplotlib.pyplot as plt
import numpy as np
import random

from sklearn.model_selection import train_test_split

hat_beta_0 = 15
hat_beta_1 = 2
errorRange = (-50, 15)
```

```python
def f(x):
    y = hat_beta_0 + hat_beta_1 * x
    error = random.uniform(*errorRange) 
    
    return y + error

values = []
for x in range(50):
    x = random.uniform(-100, 100)
    y = f(x)
    values.append([x, y])
    
X = np.array(list(zip(*values))[0])
y = np.array(list(zip(*values))[1])
plt.scatter(X, y)
plt.show()
```

```python
trainX, testX, trainY, testY = train_test_split(X, y, test_size=.33)

plt.scatter(trainX, trainY, color='b')
plt.scatter(testX, testY, color='r')
plt.show()
```

```python
meanTrainX = trainX.mean()
meanTrainY = trainY.mean()

beta_1 = sum((trainX-meanTrainX)*(trainY-meanTrainY)) / sum((trainX-meanTrainX)**2)
beta_0 = meanTrainY - beta_1 * meanTrainX
[beta_0, beta_1]
```

```python
minTrainX, maxTrainX = trainX.min(), trainX.max()

guessY_0 = beta_0 + beta_1 * minTrainX
guessY_1 = beta_0 + beta_1 * maxTrainX

plt.plot((minTrainX, maxTrainX), (guessY_0, guessY_1), color='g')
plt.scatter(trainX, trainY, color='b')
plt.scatter(testX, testY, color='r')

plt.show()
```

```python
def loss(beta_0, beta_1, X, Y):
    return sum((Y - ( hat_beta_0 + hat_beta_1 * X ))**2)

loss(beta_0, beta_1, testX, testY)
```
