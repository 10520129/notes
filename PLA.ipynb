```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
n_data = 5
```

```python
def f(x):
    return 3*x + 2

data = []
values = []
for i in range(n_data):
    x, y = np.random.normal(10, 10), np.random.normal(5, 10)
    values.append([np.array([[x], [y]]), 1])
    
X = list(zip(*values))[0]

plt.scatter(list(zip(*X))[0], list(zip(*X))[1], color='b')
data.extend(values)

values = []
for i in range(n_data):
    x, y = np.random.normal(-25, 10), np.random.normal(15, 10)
    values.append([np.array([[x], [y]]), -1])

X = list(zip(*values))[0]

plt.scatter(list(zip(*X))[0], list(zip(*X))[1], color='r')
data.extend(values)

plt.show()

data = np.array(data)
```

```python
def sign(W, X):
    return 1 if np.dot(W.T, X) > 0 else -1

W = np.zeros((3,1))
XX = list(zip(*data))[0]
minX, maxX = min(list(zip(*XX))[0]), max(list(zip(*XX))[0])

while True:
    
    error = False
    for X, y in data:
        X = np.insert(X, 0, 1, axis=0)
        if sign(W,X) != y:
            W = W + y*X
            error = True
            break
    
    if not error: break
        
    plt.scatter(list(zip(*XX))[0][n_data:], list(zip(*XX))[1][n_data:], color='r')
    plt.scatter(list(zip(*XX))[0][:n_data], list(zip(*XX))[1][:n_data], color='b')
    xs = np.linspace(minX,maxX)
    a,b = -W[1]/W[2], -W[0]/W[2]
    plt.plot(xs, a*xs + b, 'g')
    
    plt.show()
```
